```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
dataset = [['Bedsheet','Cushion','Pillow','Cushion Cover','Pillow Cover'],
           ['Mattresses','Cushion','Pillow','Cushion Cover','Pillow Cover'],
           ['Bedsheet','Curtains','Cushion Cover','Pillow Cover'],
           ['Bedsheet','Mattresses','Blankets','Cushion Cover','Pillow Cover'],
           ['Blankets','Cushion','Cushion','Cushion Cover','Bath Towels'],
           ['Floor Coverings','Curtains','Mattresses','Pillow','Cushion'],
           ['Pillow Cover','Mattresses','Bedsheet','Cushion Cover','Blankets'],
           ['Floor Coverings','Cushion','Cushion Cover','Pillow','Bedsheet'],
           ['Mattresses','Blankets','Bath Towels','Pillow','Cushion'],
           ['Bedsheet','Blankets','Pillow Cover','Floor Coverings','Cushion Cover']]
            
```

```python
te = TransactionEncoder()
```

```python
data = te.fit(dataset).transform(dataset)
```

```python
import pandas as pd
```

```python
productdata = pd.DataFrame(data,columns = te.columns_)
```

```python
productdata
```

```python
#Market basket analysis basically uses three diferent matrix
#1.Support 2. Confidence and 3. Lift.
#Support= It gives a fraction of records which contains factor the factor can be either one or multiple basically support tells us about the frequent combination of factors.

#Support = No of records per transactions containing factors (A,b or both ab)/No of total records.

#Confidence = No of records containing both a and b/No of records containing A.(Confidence tells us about how often factors a and b occur together given no of times A occurs)

#Eg - Suppose A is bread and B is butter the value of confidence will depict the number of records when bread and butter were together purchased/ When bread alone was purchased.

##Lift =It indicates the strength of the role over the random occurence of A and B . The greater the value of lift the more is the strength.
#If Lift is less than 1 it means that there is no association.

#If lift is equal to 1 this means that there is a very poor association.

#If lift is greater than 1 that means it is a good association.

#If lift is equal to 3 it means that if u buy A chances of buying B is three times .

#lift = Support of X union Y / Support of X * Support of Y
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

#Determining the frequent factors
frequent_factors = apriori(productdata,use_colnames=True,min_support=0.1,max_len=4)

#Filtering the factors on the basis of conditions
rules = association_rules(frequent_factors,metric ='lift',min_threshold=1)

#Displaying the first 10 records
print(rules.head(10).sort_values(by='confidence',ascending =False))
```

```python
#List the products which were sold min 50% times
apriori(productdata,use_colnames=True,min_support =0.5)
```

```python
#The data of the market basket analysis should be in the list format only.
```

```python
frequent_itemsets = apriori(productdata,min_support=0.6,use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x:len(x))
```

```python
frequent_itemsets[(frequent_itemsets['length']>=2) & (frequent_itemsets['support']>=0.5)]
```

```python
frequent_itemsets[(frequent_itemsets['itemsets'] == {'Cushion Cover','Bedsheet'})]
```

```python
target = '{\'Bedsheet\'}'

#Searching required strings in consequents and sorting values on confidence
results_Bedsheet = rules[rules['antecedents'].astype(str).str.contains(target,na =False)].sort_values(by='confidence',ascending =False)

#Displaying first 10 records
results_Bedsheet.head(10)
```
